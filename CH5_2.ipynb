```python
import pandas as pd
df=pd.read_csv("tshirt.csv")
df
```

```python
y=df.iloc[:,-1]
x=df.iloc[:,:2]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
nn=KNeighborsClassifier(n_neighbors=5)
model=nn.fit(x,y)
prediction=model.predict(x)
print(prediction)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y,prediction)
print(cm)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
plt.figure(figsize=(4,4))
cm_disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=["A","B"])
cm_disp.plot()
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df
```

```python
df.isna().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df
```

```python
df.isna().sum()
```

```python
import numpy as np
df['Sex']=np.where(df['Sex']=='male',1,0)
```

```python
df
```

```python
df.columns
```

```python
x=df[['Pclass','Sex', 'Age', 'SibSp','Parch','Fare']]
y=df['Survived']
```

```python
print(x.shape)
print(y.shape)
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

from sklearn.ensemble import RandomForestClassifier
rf_classifier=RandomForestClassifier(criterion='entropy',n_estimators=100,random_state=1)

rf_classifier.fit(x_train,y_train)
prediction=rf_classifier.predict(x_test)
prediction
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
cm=confusion_matrix(y_test,prediction)
print(cm)
```

```python
accuracy_score(y_test,prediction)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
chosen_tree=rf_classifier.estimators_[0]

plot_tree(chosen_tree,feature_names=x.columns,
class_names=['Not Survived','survived'],filled=True,rounded=True)
plt.show()
```

# support vector machine

```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
df
```

```python
x=df.iloc[:,:8]
y=df.iloc[:,-1]

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='rbf',C=1.0,random_state=0)   # radial basic function //  by default rbf // c=regularation  
model=classifier.fit(x_train,y_train)
y_prediction=model.predict(x_test)
y_prediction
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
cm=confusion_matrix(y_test,y_prediction)
print(cm)
```

```python
accuracy_score(y_test,y_prediction)
```

```python
recall_score(y_test,y_prediction)
```

```python
precision_score(y_test,y_prediction)
```
